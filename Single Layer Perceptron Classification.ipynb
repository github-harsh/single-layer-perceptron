                                                 Single Layer Perceptron
                   
                   Take the know data pass it to activation funciton -> gives guess +1 or -1
                   We also have the answer, so we can compute error -> answer - guess.
                   We want to find optimal weigth -> W0 + delta W0
                   
                   The possible errors will be as follows:
                   
                   Output               Desired              Error
                     +1                   -1                  +2
                   ==============================================
                     +1                   +1                   0
                   ==============================================
                     -1                   -1                   0
                   ==============================================
                     -1                   +1                  -2

```python
import numpy as np;
import matplotlib.pyplot as plt;

class Perceptron:
    learningRate = 0.0001
    weights=np.random.uniform(-1,1,2)
    print ("Initial weights :",weights)
    xAxis = np.random.randint(0,100,100)
    yAxis = np.random.randint(0,100,100)
    inputs=[xAxis , yAxis]
    label = [None] * 100
    trainingLabel = [None] * 100
    finalTrainedLabel = [None] * 100
    print len(label)
    count = 0
   
    
    #Assign the labels to inputs for training
    def assignLabelsToCreateTrainingData(self):
        for i in range(0,100):
            if self.xAxis[i] >= self.yAxis[i]:
                self.label[i] = 1
            if self.xAxis[i] < self.yAxis[i]:
                self.label[i] = -1
        self.show()
    
    #Calculate the sum of weighted inputs and error
    def assignLabelsByPassingThroughNetwork(self):
       
        result = 0
        __i = 0
        for i in range(0, 100):
            __i = i
            IW1 = self.inputs[0][i] * self.weights[0]
            IW2 = self.inputs[1][i] * self.weights[1]
            result = IW1 + IW2 
            #print result
            if result >= 0:
                self.trainingLabel[i] = 1
            else:
                self.trainingLabel[i] = -1
                
            error = self.label[i] - self.trainingLabel[i]                 
            self.train(error,__i)
        self.guessOfNetwork()
        
        self.count += 1
        print "count",self.count
        if self.count == 2:            
            self.createAndPlotTrainedNetworkData()
            #Call to create array of labels based of the trained network


    def train(self,error,__i):
        #Tune all the weights for the particular error
        for i in range(0,2):
            self.weights[0] += self.learningRate * error * self.inputs[0][__i]
            self.weights[1] += self.learningRate * error * self.inputs[1][__i]
        
        
    
    def createAndPlotTrainedNetworkData(self):
        print "here"
        result = 0
        for i in range(0, 100):
            IW1 = self.inputs[0][i] * self.weights[0] 
            IW2 = self.inputs[1][i] * self.weights[1]
            print "y :",self.inputs[1][i], "weight :",self.weights[1]," multiplication :",self.inputs[1][i] * self.weights[1]," actual : ",IW2
            print self.inputs[1][i] * self.weights[1]
            result = IW1 + IW2
            #print "x :",self.inputs[0][i], "y :",self.inputs[1][i], "IW1 : ",IW1, " IW1 : ",IW2," result :" , result
            if result > 0:
                self.finalTrainedLabel[i] = 1
            else:
                self.finalTrainedLabel[i] = -1
        
        print ("Final weights :",self.weights)
        for i in range(0,100):
            if self.finalTrainedLabel[i] == 1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'ro')
            if self.finalTrainedLabel[i] == -1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'bo')
        plt.plot([0, 100], [0, 100])
        plt.axis([0, 100, 0, 100])
        plt.title("Result of Perceptron")
        plt.show()
        
        
    def show(self):
        for i in range(0,100):
            if self.label[i] == 1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'ro')
            if self.label[i] == -1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'bo')
        plt.plot([0, 100], [0, 100])
        plt.axis([0, 100, 0, 100])
        plt.title("Training Data")
#         plt.show()
        
    def guessOfNetwork(self):
        for i in range(0,100):
            if self.trainingLabel[i] == 1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'ro')
            if self.trainingLabel[i] == -1:
                plt.plot(self.xAxis[i], self.yAxis[i], 'bo')
        plt.plot([0, 100], [0, 100])
        plt.axis([0, 100, 0, 100])
        plt.title("Guess of Network")
        plt.show()
    
   
        
        

# Untrained Data 
obj = Perceptron()
# obj.createAndPlotTrainedNetworkData()
obj.assignLabelsToCreateTrainingData()
obj.assignLabelsByPassingThroughNetwork()
obj.assignLabelsByPassingThroughNetwork()

#Test the inputs and adjust the weights

#SOWI = obj.train()

#print SOWI
 
```
